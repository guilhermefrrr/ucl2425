```python
!pip install requests beautifulsoup4 --quiet

import requests
from bs4 import BeautifulSoup

url = "https://fbref.com/en/comps/8/schedule/Champions-League-Scores-and-Fixtures"
response = requests.get(url)
response.raise_for_status()  # Raise an exception for bad status codes
soup = BeautifulSoup(response.content, 'html.parser')

match_report_links = []
for a_tag in soup.find_all('a', href=True, text='Match Report'):
    match_report_links.append(a_tag['href'])

base_url = "https://fbref.com"
complete_urls = [base_url + link for link in match_report_links]
```

# Data processing and results

```python
for url in complete_urls:
    print(url)
```
