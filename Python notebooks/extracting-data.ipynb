### *'check_429_error'* function

```python
# Implementing error handling function for HTTP Error 429
def check_429_error(response_or_exception):
    """
    Lida com o erro 429 (Too Many Requests) em diferentes contextos.

    Args:
        response_or_exception: O objeto response da requisição ou a exceção HTTPError.

    Raises:
        requests.exceptions.HTTPError: Se o erro 429 for encontrado.
    """
    if isinstance(response_or_exception, requests.models.Response):
        # Handles successful responses but checks for a 429 status code
        response = response_or_exception
        if response.status_code == 429:
            print("HTTP Error 429 encountered. Stopping execution")
            raise requests.exceptions.HTTPError(response=response)
    elif isinstance(response_or_exception, requests.exceptions.HTTPError):
        # Handles exceptions specifically of type HTTPError and checks for a 429 status code
        err = response_or_exception
        if err.response.status_code == 429:
            print("HTTP Error 429 encountered. Stopping execution")
            raise
    else:
        # Handles other types of exceptions or unexpected objects
        print("Unexpected error or object type:", type(response_or_exception))
```

#### Detalhamento da função *'check_429_error'*


A função check_429_error é como um "porteiro" que verifica se o site está sobrecarregado (erro 429) antes de permitir que seu programa acesse as informações.

Analogia:

**Imagine que o porteiro precisa lidar com dois tipos de visitantes:**

1. Visitantes diretos: Aqueles que chegam diretamente ao site, sem passar por intermediários.
2. Visitantes com encaminhamento: Aqueles que chegam ao site através de um guia turístico (no caso do seu código, o pd.read_html).

**Como o porteiro lida com cada tipo de visitante:**

1. Visitantes diretos: O porteiro pode verificar diretamente se eles possuem o erro 429, observando se eles estão apresentando algum sinal de impaciência ou frustração (como o código de status 429 na resposta do site).
2. Visitantes com encaminhamento: O porteiro precisa primeiro conversar com o guia turístico para entender se houve algum problema durante o caminho. O guia turístico pode relatar que o visitante está com o erro 429, mas essa informação estará encapsulada em um relatório (a exceção HTTPError). O porteiro precisa então analisar o relatório para confirmar se o visitante realmente possui o erro 429.

**Por que a diferença?**

A diferença na forma como o porteiro lida com cada tipo de visitante se deve ao fato de que a informação sobre o erro 429 é apresentada de maneiras diferentes:

1. Visitantes diretos: O porteiro pode observar diretamente o erro 429.
2. Visitantes com encaminhamento: O porteiro precisa obter a informação sobre o erro 429 através do guia turístico e do relatório que ele fornece.
Conclusão:

Dentro da função check_429_error, você precisa tratar o erro 429 de formas diferentes porque a informação sobre o erro é apresentada de maneiras diferentes, dependendo de como a requisição foi feita:

Se ela foi feita diretamente com requests.get, como acontece na primeira requisição do código abaixo, você tem acesso direto à resposta do site e pode verificar o código de status usando response.status_code.

Se ela foi feita através de pd.read_html, como acontece na segunda requisição do código abaixo, você recebe uma exceção HTTPError e precisa acessar o código de status através do atributo err.response.status_code.

A função check_429_error que criamos não possui uma instrução return explícita. Isso significa que, se a condição do if ('if response.status_code == 429:' ou 'if err.response.status_code == 429:') não for satisfeita, a função retornará implicitamente None.

Em Python, quando uma função termina sem encontrar uma instrução return explícita, ela retorna automaticamente None. Isso é considerado um comportamento padrão da linguagem.

Na função, o else imprime uma mensagem indicando que recebeu um tipo de objeto ou exceção inesperado. Ou seja não é nem um visitante direto, nem um visitante com encaminhamento.


### *'extract_squad_names'* function

```python
# Extracting squad names from the page title. Exampe: 'Milan vs. Liverpool Match Report – Tuesday September 17, 2024'
def extract_squad_names(soup):
    title_container = soup.find(id="content", role="main")
    title = title_container.find('h1')
    padrao = r'(.*?)\s+vs\.\s+(.*?)\s+Match Report'
    match = re.search(padrao, title.text)

    if match:
        squad1 = match.group(1)
        squad2 = match.group(2)
        return squad1, squad2
```

### '*extract_squad_ids*' function

```python
# Extracting squad IDs. THe iteration below identifies the ID of each of the two 'summary' tables and based on them the club ID is extracted
def extract_squad_ids(soup, squad1, squad2):
    squad_ids = [re.findall(r'stats_([^_]+)_summary', tag['id'])[0] for tag in soup.find_all('table') if tag.has_attr('id') and re.match(r'stats_.+_summary', tag['id'])]
    ids = {squad1: squad_ids[0], squad2: squad_ids[1]}  # Creating a dictionary with squads names and IDs
    return ids
```

### *'extract_last_word'* function

```python
# Extracting the last word from a column name
def extract_last_word(column_name):
    last_word = re.sub(r'.*_', '', column_name)
    return last_word
```

### *'flatten_column_names'* function

```python
# Converting multi-level column names (also called hierarchical columns) into single-level column names
def flatten_column_names(df):
  # Joining the levels with an underscore
  df.columns = ['_'.join(col).strip() for col in df.columns.values]

  # Cleaning column names, removing "Unnamed"
  df.columns = [extract_last_word(col) if 'Unnamed' in col else col for col in df.columns]

  return df
```

### *'merge_dfs'* function

```python
def merge_dfs(df, merge_keys):
  # Initializing the merged DataFrame with the first DataFrame from the dictionary
  df_merged = dfs[list(dfs.keys())[0]]  # Using the first DataFrame as a base. In this case, it's the 'summary' DataFrame, the first one iterated

  # Iterating and merging the other DataFrames, excluding duplicate columns
  for df_id in list(dfs.keys())[1:]:
    df = dfs[df_id]  # Current DataFrame to be merged

    # Selecting the columns to be used, using only those that are not yet present in the merge
    cols_to_use = [col for col in df.columns if col not in df_merged.columns or col in merge_keys]

    # Merging the tables based on the specified keys and selected columns, using 'outer' join"
    df_merged = pd.merge(df_merged, df[cols_to_use], on=merge_keys, how='outer')

  return df_merged
```

### *'clean_df'* function

```python
def clean_df(df):
  # Eliminating rows containing a number followed by 'players' in the 'Player' column. By doing this, we remove the row that aggregates data for all players who played for the team
  df_cleaned = df_merged[~df_merged['Player'].str.contains(r'\d+\sPlayers', regex=True)]

  # List of columns to be removed
  colums_to_remove = ['Total_Cmp', 'Total_Att', 'Total_Cmp%', 'Ast', 'xAG', 'PrgP', 'Att', 'Tackles_Tkl', 'Blocks_Blocks', 'Touches_Touches', 'Touches_Live', 'Crs']

  # Dropping columns from the df_merged DataFrame
  df_cleaned = df_cleaned.drop(columns=colums_to_remove, errors='ignore')

  # List of strings to be checked
  special_strings = ['Carries_Carries', 'Receiving', 'Performance', 'Expected', 'SCA']

  # Renaming columns from the 'special_strings' variable
  new_columns = [extract_last_word(col) if any(string in col for string in special_strings) else col for col in df_cleaned.columns]
  df_cleaned.columns = new_columns

  # Extracting country codes from 'Nation' column
  df_cleaned['Nation'] = df_cleaned['Nation'].str.split().str.get(1)

  # Converting the '#' column values to integers
  df_cleaned['#'] = df_cleaned['#'].astype(int)

  # Splitting the 'Age' column into two using the hyphen as a separator
  df_cleaned[['Age_Years', 'Age_Days']] = df_cleaned['Age'].str.split('-', expand=True)

  # Converting the new columns to numeric, handling errors with NaN
  df_cleaned['Age_Years'] = pd.to_numeric(df_cleaned['Age_Years'], errors='coerce')
  df_cleaned['Age_Days'] = pd.to_numeric(df_cleaned['Age_Days'], errors='coerce')

  # Removing the original 'Age' column
  df_cleaned = df_cleaned.drop(columns=['Age'])

  # Reordering the columns to maintain the original position of 'Age'
  age_index = df_cleaned.columns.get_loc('Nation')  # Finding the position of the 'Nation' column
  df_cleaned.insert(age_index + 1, 'Age_Years', df_cleaned.pop('Age_Years'))  # Inserting 'Age_Years' after 'Nation'
  df_cleaned.insert(age_index + 2, 'Age_Days', df_cleaned.pop('Age_Days'))  # Inserting 'Age_Days' after 'Age_Years'

  return df_cleaned
```

### *'enrich_df'* function

```python
def enrich_df(df):
  # Adding the 'Match_ID' column in the first position
  df_enriched = df_cleaned.copy()

  if match_ID:  # Checks if match_ID is not empty (evaluates to True)
    df_enriched.insert(0, 'Match_ID', match_ID)

  if MD:  # Checks if MD is not empty (evaluates to True)
    current_MD = ((match_ID - 1) // matches_per_MD) + 1
    df_enriched.insert(0, 'MD', current_MD)

  # Adding the 'Squad' column to the right of the 'Player' column
  player_col_index = df_enriched.columns.get_loc('Player')
  df_enriched.insert(player_col_index + 1, 'Squad', squad_name)

  return df_enriched
```

```python
# Iterating over match URLs
for i, match_url in enumerate(match_urls):
    start_time = time.time()  # Recording the start time

    response = requests.get(match_url) # Making the HTTP request for the server to send the page content
    check_429_error(response)  # Calls 'check_429_error' function with response object

    # Transforming this HTML code into a BeautifulSoup object, which will be used to extract information from the page
    soup = BeautifulSoup(response.content, 'html.parser')

    squad1, squad2 = extract_squad_names(soup)

    ids = extract_squad_ids(soup, squad1, squad2)

    # Iterating over squads IDs
    for squad_name, squad_ID in ids.items():
        # Dictionary to store DataFrames
        dfs = {}

        # Iterating over the squad's DataFrames
        for df_id in df_ids:
            # Making the HTTP request to read each DataFrame individually and checking the 429 error
            try:
                df = pd.read_html(match_url, attrs={"id": f"stats_{squad_ID}_{df_id}"})[0]
            except requests.exceptions.HTTPError as err:
                check_429_error(err)  # Calls 'check_429_error' function with 'err' exception

            df = flatten_column_names(df)

            # Storing the DataFrame in the dictionary
            dfs[df_id] = df

            time.sleep(5) # Waits 5 seconds before the next request, which will extract data for the team's next DataFrame

        df_merged = merge_dfs(df, merge_keys)

        df_cleaned = clean_df(df_merged)

        df_enriched = enrich_df(df_cleaned)

        all_data.append(df_enriched)

        time.sleep(10) # Waits 10 seconds before the next request, which will extract data for the next squad

    end_time = time.time()  # Recording the end time
    elapsed_time = end_time - start_time  # Calculating the elapsed time for processing each match's data
    print(f"Match {match_ID} data ({squad1} vs {squad2}) successfully extracted and processed in {elapsed_time:.2f} seconds")
    if match_ID:
      match_ID += 1 # Updating match_ID

df_final = pd.concat(all_data, ignore_index=True)
```
